## Predicting Salaries for Data Scientist Jobs
###  Summary

The aim of this project was to investigate what industry factors influence the salary of data scientist jobs in the market across USA. Due to natural variances in job salaries, instead of trying to predict the actual salary, predicting a range for salary i.e. above or below median could be a useful and insightful alternative.
Information on data scientist jobs was collected from Indeed.com resulted in predicting senioroty and the location of a job (being in Chicago,San Jose, San Francisco Bay Area) to be the most important factors in contributing to classifying a job above or below median.

### Introduction

“Data Scientist” has been ranked the hottest job in the US market for the past few years and data scientist are in demand. It is really challenging for companies to find a data scientist with right skills to hire. Offering competitive salaries would result in higher  success rates in landing a good data scientist. Therefore, knowing what factors mainly influence the salary of a data scientist job is critical to be able to offer right and competitive salaries. 

### Methods
For this project it was required to get information on Data Scientist jobs available in different cities such as the salary, the location and the job title from Indeed.com. We would like to find out what factors most directly impact salaries. The objective is to predict whether a job posting would have a salary above the median or below median based on factors and features we considered in our model. Therefore, this problem becomes a classification problem.
The first step is  to get the data from the indeed.com website. The job searching website of indeed.com can be scraped by providing the urls.  For each cities among 19 selected cities (NewYork, Chicago, San Francisco, Austin, Seattle, Los Angeles,Philadelphia, Atlanta, Dallas, Pittsburgh, Portland, Phoenix, Denver, Houston, Miami, San Jose,Palo Alto, Mountain View, Washington,DC) in USA, I scraped 10 pages of results for each different ranges of estimated salaries (75,000, 90,000, 105,000, 120,000, 135,000 and over 150,000). The estimated salaries were used in cases where not enough data had been found with salary information. The info on the job title, location, title, company, and the link to the summary of the job description is saved in a csv file to avoid the need to scrape the website in multiple occasions.


```python
# importing required libraries
import pandas as pd  
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
from sklearn.model_selection import cross_val_score,StratifiedKFold,cross_val_predict
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,roc_auc_score,roc_curve, auc
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV,train_test_split 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer,HashingVectorizer,TfidfVectorizer
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
# making the url address list
cities = ['New+York', 'Chicago', 'San+Francisco', 'Austin', 'Seattle', 
    'Los+Angeles', 'Philadelphia', 'Atlanta', 'Dallas', 'Pittsburgh', 
    'Portland', 'Phoenix', 'Denver', 'Houston', 'Miami','San+Jose','Palo+Alto','Mountain+View', 'Washington,+DC']
salaries = ['$75,000','$90,000','$105,000','$120,000','$135,000', '$150,000']
salary_range = 15000/2
url_template=[]
for city in cities:
    for salary in salaries:
        for i in range(0,10):
            url_template.append('https://www.indeed.com/jobs?q=%22data+scientist%22'+'+'+salary+'&l='+ city+'&start='+str(i*10)+'&pp=')
```

```python
# Scraping the urls
indeed = pd.DataFrame(columns=["company","location","title",'salary_original', 'salary_estimated','link'])
# loop through each entry. row result values are typed differently with spaces, to get all of them used regex having result at the end
for item in url_template:
    html = requests.get(item)
    b = BeautifulSoup(html.text)
    for entry in b.find_all('div', {'class':re.compile(r'result$')}):
        # grab the company name
        company = entry.find('span', {'class': 'company'}).text.replace('\n','')
        # grab the location
        location = entry.find('span', {'class': 'location'}).text
        
        # grab the job title
        title = entry.find('a', {'data-tn-element':'jobTitle'}).text
        # try to find the salary
        try:
            temp = entry.find('nobr').text
        
            if temp:
                temp = temp.replace('a year', '')
                temp = temp.replace(',','')
                temp = temp.replace('$','')
                salary_original = temp
        except:
            salary_original = 'NA'
        # try to estimate the salary for those with no salary 
        try:
            temp = entry.find('nobr').text
        
            if temp:
                salary_estimated = "Not Required"
        except:
            # getting the salary from the salary range in the url + half of the range from this salary and the next salary range
            salary_estimated =''.join(re.findall(r'\$(\d*\,\d*)', item))
            salary_estimated = salary_estimated.replace(',','')
            if float(salary_estimated) < 150000:
                salary_estimated =float(salary_estimated)+ salary_range
            else:
                # to average the salaries above 150000 and 190000
                salary_estimated =float(salary_estimated)+ 20000

        # grab the link
        link = entry.find('a', {'class':'turnstileLink'}).get('href')
        
    #Add the company name, location, job title , salary and link to the dataframe
        indeed.loc[len(indeed)]=[company, location, title, salary_original, salary_estimated,link]
```

```python
#saving the data as csv file, so I do not need to scrape the website everytime
indeed.to_csv('indeed_data.csv',encoding ='utf-8',index_label=False )
```

```python
#Reading the scraped the data into a dataframe
data = pd.read_csv('indeed_data.csv')
```

After dropping the duplicate entries in the dataframe, I ended up with 10910 (down from 15113). Tha data needs further manipulations such as cleaning the locations strings, converting the salary values to numbers. 

```python
#removing the rows with hourly,monthly or weekly salary input
data = data[~((data['salary_original'].str.contains('an hour')) | (data['salary_original'].str.contains('a month')) |(data['salary_original'].str.contains('a week')))]
```

```python
# putting the average value instead of a range for salary for the ones with original salary othrewise the estimated salary

data['salary'] = np.where(data['salary_estimated'].str.contains("Not Required")== True,data['salary_original'], data['salary_estimated'])
data['salary'] = np.where(data['salary'].str.contains("-")== True,[sum([float(i) for i in x.split(' - ')])/2 for x in data['salary']], data['salary'])
```

```python
# make the salary column as integer type
data['salary']=data['salary'].astype(float)
```

```python
#getting rid of numbers, details in parantheses and white spaces, and the entries with United States as location

data['location'] = np.where(data['location'].str.contains(r'\d+')== True,[re.sub(r'\d+', '',x) for x in data['location']],data['location'])
data['location'] = np.where(data['location'].str.contains(r'\([^)]*\)')== True,[re.sub(r'\([^)]*\)', '',x) for x in data['location']],data['location'])
data['location'] = [" ".join(x.split())for x in data['location']]
data=data[data['location']!='United States']
```

First, I am using only the data enteries with original salary posted on the website. Only 124 data point had the salary posting. A column named "pay" was made which has a value of 1 for jobs paying more than median and 0 for jobs with salaries below median. The medianof data scietist jobs with salary posted on indeed is $150000.

## Fitting random forest and knn classifiers  having location as feature.

```python
# Random forest with cross validation score using locations as features
y = data_original_salary['pay']
X = pd.get_dummies(data_original_salary['location'])
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=33)
rf = RandomForestClassifier(n_jobs=-1, random_state=10)
scores = cross_val_score(rf, X, y, cv=skf)
y_pred = cross_val_predict(rf, X, y, cv=skf)

conmat = np.array(confusion_matrix(y, y_pred, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['1', '0'],
                         columns=['predicted_1','predicted_0'])


print(scores)
print('mean of cv scores:',scores.mean())
print('confusion_matrix:')
print(confusion)
print('classification_report:')
print(classification_report(y, y_pred))
print('accuracy_score:')
print(accuracy_score(y, y_pred))
```

```python
# KNN with cross validation score by using location only
y = data_original_salary['pay']
X = pd.get_dummies(data_original_salary['location'])
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=33)
kn = KNeighborsClassifier(n_jobs=-1)
scores = cross_val_score(kn, X, y, cv=skf)
y_pred = cross_val_predict(kn, X, y, cv=skf)
conmat = np.array(confusion_matrix(y, y_pred, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['1', '0'],
                         columns=['predicted_1','predicted_0'])


print(scores)
print('mean of cv scores:',scores.mean())
print('confusion_matrix:')
print(confusion)
print('classification_report:')
print(classification_report(y, y_pred))
print('accuracy_score:')
print(accuracy_score(y, y_pred))
```

Using the default values of knn and random forest models reslulted in 0.61 and 0.67 accuracy scores accordingly.

## Using key words in the title and locations in random forest, logestic regression and KNN models
As the next step, key words  such as Senior, Engineer, Lead, VP and director was added as features. Then again random forest and knn models were used to classify the data.



```python
#creating a column senior: 1 for jobs that have sr. or senior in the job title (ignore case sensitive, flags = re.I) and 0 elsewhere 
data_original_salary['senior'] = np.where(((data_original_salary['title'].str.contains("sr.", flags = re.I)== True) | (data_original_salary['title'].str.contains("senior", flags = re.I)== True)),1, 0)

#creating a column lead: 1 for jobs that have lead in the job title (ignore case sensitive, flags = re.I) and 0 elsewhere 
data_original_salary['lead'] = np.where((data_original_salary['title'].str.contains("lead", flags = re.I)== True),1, 0)

#creating a column engineer: 1 for jobs that have engineer in the job title (ignore case sensitive, flags = re.I) and 0 elsewhere 
data_original_salary['engineer'] = np.where((data_original_salary['title'].str.contains("engineer", flags = re.I)== True),1, 0)

#creating a column director: 1 for jobs that have director in the job title (ignore case sensitive, flags = re.I) and 0 elsewhere 
data_original_salary['director'] = np.where((data_original_salary['title'].str.contains("director", flags = re.I)== True),1, 0)

#creating a column VP: 1 for jobs that have VP in the job title and 0 elsewhere 
data_original_salary['VP'] = np.where((data_original_salary['title'].str.match("VP")== True),1, 0)
```

```python
# Random forest with cross validation by using location and key words in the title
y = data_original_salary['pay']
X = pd.get_dummies(data_original_salary['location'])
X1 =data_original_salary[['senior','lead','engineer','director','VP']]
X = pd.concat([X,X1], axis =1)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=33)
#max_features defualt is sqrt numbr of features(sqrt(34))
rf = RandomForestClassifier(n_jobs=-1, random_state=10)

scores = cross_val_score(rf, X, y, cv=skf)
y_pred = cross_val_predict(rf, X, y, cv=skf)
conmat = np.array(confusion_matrix(y, y_pred, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['1', '0'],
                         columns=['predicted_1','predicted_0'])


print(scores)
print('mean of cv scores:',scores.mean())
print('confusion_matrix:')
print(confusion)
print('classification_report:')
print(classification_report(y, y_pred))
print('accuracy_score:')
print(accuracy_score(y, y_pred))
```

```python
# logestic regression with cross validation by using location and key words in the title

y = data_original_salary['pay']
X = pd.get_dummies(data_original_salary['location'])
X1 =data_original_salary[['senior','lead','engineer','director','VP']]
X = pd.concat([X,X1], axis =1)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=33)
#max_features defualt is sqrt numbr of features(sqrt(34))
lr = LogisticRegression(n_jobs=-1, random_state=10)

scores = cross_val_score(lr, X, y, cv=skf)
y_pred = cross_val_predict(lr, X, y, cv=skf)
conmat = np.array(confusion_matrix(y, y_pred, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['1', '0'],
                         columns=['predicted_1','predicted_0'])


print(scores)
print('mean of cv scores:',scores.mean())
print('confusion_matrix:')
print(confusion)
print('classification_report:')
print(classification_report(y, y_pred))
print('accuracy_score:')
print(accuracy_score(y, y_pred))
```

```python
# KNN with cross validation score by using locations as key words in title
y = data_original_salary['pay']
X = pd.get_dummies(data_original_salary['location'])
X1 =data_original_salary[['senior','lead','engineer','director','VP']]
X = pd.concat([X,X1], axis =1)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=33)
kn = KNeighborsClassifier(n_jobs=-1)
scores = cross_val_score(kn, X, y, cv=skf)
y_pred = cross_val_predict(kn, X, y, cv=skf)

conmat = np.array(confusion_matrix(y, y_pred, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['1', '0'],
                         columns=['predicted_1','predicted_0'])


print(scores)
print('mean of cv scores:',scores.mean())
print('confusion_matrix:')
print(confusion)
print('classification_report:')
print(classification_report(y, y_pred))
print('accuracy_score:')
print(accuracy_score(y, y_pred))

```

## Improving models by grid search over model parameters

For both random forest and knn adding the key words in the title reduced the accuracy of the model to 0.60. As the next step, using grid search I tried to improve the model accuracy.

```python
#Doing grid search on max_features in random forest 

y = data_original_salary['pay']
X = pd.get_dummies(data_original_salary['location'])
X1 =data_original_salary[['senior','lead','engineer','director','VP']]
X = pd.concat([X,X1], axis =1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 24)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=33)
#max_features defualt is sqrt numbr of features(sqrt(40))
rf = RandomForestClassifier(n_jobs=-1, random_state=10)

grid = GridSearchCV(rf, cv =skf, param_grid={'max_features': [.2,.5,1]} )
grid_fit = grid.fit(X_train, y_train)
print(grid_fit.best_params_)
y_pred = grid_fit.predict(X_test)
conmat = np.array(confusion_matrix(y_test, y_pred, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['1', '0'],
                         columns=['predicted_1','predicted_0'])

print('confusion_matrix:')
print(confusion)
print('classification_report:')
print(classification_report(y_test, y_pred))
print('accuracy_score:')
print(accuracy_score(y_test, y_pred))
```

```python
#Doing grid search on number of neighbors in knn model

y = data_original_salary['pay']
X = pd.get_dummies(data_original_salary['location'])
X1 =data_original_salary[['senior','lead','engineer','director','VP']]
X = pd.concat([X,X1], axis =1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 24)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=33)

kn = KNeighborsClassifier(n_jobs=-1)

grid = GridSearchCV(kn, cv =skf, param_grid={'n_neighbors': [5,8,10,12,20]} )
grid_fit = grid.fit(X_train, y_train)
print(grid_fit.best_params_)
y_pred = grid_fit.predict(X_test)
conmat = np.array(confusion_matrix(y_test, y_pred, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['1', '0'],
                         columns=['predicted_1','predicted_0'])

print('confusion_matrix:')
print(confusion)
print('classification_report:')
print(classification_report(y_test, y_pred))
print('accuracy_score:')
print(accuracy_score(y_test, y_pred))
```

Grid search improved the accuracies of both random forest and knn from 0.6 to 0.64 and 0.68 respectively. The best parameters for the random forest are max_features of 0.5. For the knn model n_neighbors = 12 resulted in best accuracy. Even with the gridsearch, the random forest model which only included the locations and not the keywords in the title showed better accuracy compared to the one which included the keywords of the title.

The precision of the model is a key factor. We want to be sure if we tell a someone that a job is paying above median i.e. TP but we are less worried about cases that we predict below median salary and they actually have above median salary i.e. FN. So, we can change the probability threshold to achieve higher precision. Instead of threshold of 0.5 the probability threshold of 0.75 was chosen which resulted in precision of 0.83 for the random forest model on predicting a job with above median salary while reducing the model accuracy to 0.6 from 0.64.


```python
#Improving the precision of the model (reducing the false positive) to improve precision of the model 

y = data_original_salary['pay']
X = pd.get_dummies(data_original_salary['location'])
X1 =data_original_salary[['senior','lead','engineer','director','VP']]
X = pd.concat([X,X1], axis =1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 24)
rf = RandomForestClassifier(max_features = 0.5, n_jobs=-1, random_state=10)
rf_fit = rf.fit(X_train,y_train)
y_pred = rf_fit.predict(X_test)

y_pred2=[]
probabilities = rf_fit.predict_proba(X_test)

#chaniging the probability for which one category is chosen vs another
for i in range(0,len(probabilities)):
    if probabilities[i][1] >= 0.75:
        y_pred2.append(1)
    else:
        y_pred2.append(0)
conmat = np.array(confusion_matrix(y_test, y_pred2, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['1', '0'],
                         columns=['predicted_1','predicted_0'])

print('confusion_matrix:')
print(confusion)
print('classification_report:')
print(classification_report(y_test, y_pred2))
print('accuracy_score:')
print(accuracy_score(y_test, y_pred2))
```

```python
# plotting the ROC-AUC curve for random forest model
y_score=[]
for i in range(0,len(probabilities)):
    y_score.append(probabilities[i][1])


FPR = dict()
TPR = dict()
ROC_AUC = dict()

# For class 1, find the area under the curve
FPR[1], TPR[1], _ = roc_curve(y_test, y_score)
ROC_AUC[1] = auc(FPR[1], TPR[1])

# Plot of a ROC curve for class 1 
plt.figure(figsize=[5,5])
plt.plot(FPR[1], TPR[1], label='ROC curve (area = %0.2f)' % ROC_AUC[1], linewidth=4)
plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('Receiver operating characteristic for salary ', fontsize=12)
plt.legend(loc="lower right")
plt.savefig('ROC_AUC.png')
plt.show()

```

```python
# feature importance for random forest
y = data_original_salary['pay']
X = pd.get_dummies(data_original_salary['location'])
X1 =data_original_salary[['senior','lead','engineer','director','VP']]
X = pd.concat([X,X1], axis =1)

# using the max_features from the gridsearch
rf = RandomForestClassifier(max_features = 0.5, n_jobs=-1, random_state=10)
rf_fit = rf.fit(X,y)
importances = rf_fit.feature_importances_
# calculate the standard deviation of feature importances by looping over the trees in the random forest
# 

std = np.std([tree.feature_importances_ for tree in rf_fit.estimators_], axis=0)


indices = np.argsort(importances)[::-1]
feature_names = X.columns

# Plot the feature importances of the forest
plt.figure()
#plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
       color="blue", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), feature_names[indices], rotation=90)
plt.xlim([-1, X.shape[1]])
plt.savefig('Feature_Importance.png')
plt.show()
```

## Using the job summary to predict salary

The next step in the project is to add the words from the job summary to the features. for getting the summary of a job from indeed, you are directed to either an indeed template website  or the job posting on the company's website. Each company has its own html structure and therefore it is hard to find tags and labels regarding the summary. However, the full job posting that are located on indeed website, follow a same template. So, I only used those job postings that have their summary in indeed format to extract words from. The overlap between the jobs with original salary posted on indeed and having the format of indeed for the posting is very small (only 30 points). Therefore, once I used the summary text in the model with only the data that had both the original salary and summary  in indeed format and another time on more data points with their salaries estimated from indeed and their summary text.

```python
#Scraping for job summaries from the links having the summary of the job
# Getting the summary for the jobs which has the salary posted
i=0
summary_list=[]
for item in data_original_salary['link']:
    
    html = requests.get('https://www.indeed.com'+item)
    b = BeautifulSoup(html.text,"lxml")
    i = i+1
    print(i)
    summary =''
    for entry in b.find_all('span', {'id': 'job_summary', 'class': 'summary'}):
        # grab the company name

        try:
            temp = entry.text
            if temp:
                summary = temp+summary
        except:
            summary = 'NA'
    summary_list.append(summary)
```

```python
#Adding the summary text to the dataframe
data1 = data_original_salary
data1['summary'] = summary_list
```

```python
# saving the dataframe to a csv file
data1.to_csv('indeed_salary_summary.csv',encoding ='utf-8',index_label=False )
```

## Random forest classifier and summary text as feature

## Using summary text as predictor on data which has either the original salary posted or the salary was estimated 

```python
#readiing the data 1000 samples with their summary
data_summary = pd.read_csv('indeed_data1.csv')
```

```python
median = data_summary.salary.median()
data_summary['pay'] = np.where(data_summary['salary'] >= median, 1,0)
```

```python
# Using Tfidf to produce features from the text of the summary and then applying random forest to predict the salary

y = data_summary['pay']
X = data_summary['summary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 24)

cvec = TfidfVectorizer(stop_words='english')
cvec.fit(X_train)
X_train  = pd.DataFrame(cvec.transform(X_train).todense(), columns=cvec.get_feature_names())
#y_train = data_train['target']

X_test = pd.DataFrame(cvec.transform(X_test).todense(), columns=cvec.get_feature_names())
rf = RandomForestClassifier(n_jobs=-1, random_state=10)
rf_fit = rf.fit(X_train,y_train)
importances = rf_fit.feature_importances_
y_pred = rf_fit.predict(X_test)
y_pred2=[]
probabilities = rf_fit.predict_proba(X_test)

#chaniging the probability for which one category is chosen vs another
for i in range(0,len(probabilities)):
    if probabilities[i][1] >= 0.75:
        y_pred2.append(1)
    else:
        y_pred2.append(0)
conmat = np.array(confusion_matrix(y_test, y_pred2, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['1', '0'],
                         columns=['predicted_1','predicted_0'])

print('confusion_matrix:')
print(confusion)
print('classification_report:')
print(classification_report(y_test, y_pred))
print('accuracy_score:')
print(accuracy_score(y_test, y_pred))
print('auc_score:')
print(roc_auc_score(y_test, probabilities[:,1]))
print('confusion_matrix:')
print(confusion)
print('classification_report:')
print(classification_report(y_test, y_pred2))
print('accuracy_score:')
print(accuracy_score(y_test, y_pred2))
print('auc_score:')
print(roc_auc_score(y_test, probabilities[:,1]))

feature_importances = pd.DataFrame(rf_fit.feature_importances_,
                                   index = X_train.columns,
                                    columns=['importance']).sort_values('importance',
                                                                        ascending=False)
```

### Discussion / Interpretation of Results
In the absence of any models, there was a 50% chance of correctly guessing whether a job has above median salary or below median salary. In this case, the accuracy of predicting the right category of salary for a job is 0.5. Implementing a model increased the accuracy by 0.18 meaning that in 68% of times you can determine whether a job is paying below median or above median correctly. 

Overestimating salaries could result in client dissatisfaction. Estimating a job as below median paying job and ending up with above average salary is less harmful than the opposite. So, it is important to avoid overestimating the salaries. Utilizing these models, one can predict whether a job is paying above median with 0.83 precision. 

However, one should keep in mind that there is a possibility that only companies offering jobs with higher salaries or goverment sector post their salary inforamation on indeed.com and this could result bias in our models.
